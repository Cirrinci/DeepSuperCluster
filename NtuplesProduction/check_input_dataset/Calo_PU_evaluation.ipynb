```python
import matplotlib.pyplot as plt
from matplotlib import colors
import numpy as np
import pandas as pd
```

```python
import mplhep as hep
plt.style.use(hep.style.CMS)
```

```python
df = pd.read_csv("check_pu_clusters_dataset.csv", sep=";")
```

```python
plt.figure(figsize=(9,8), dpi=100)
mask = df.n_calopu > 0.
d = df[mask]
plt.scatter(d.simen_pu_frac, d.simen_sig_frac, c=d.xtals)
plt.xlabel("simEnergy PU / Cl. energy")
plt.ylabel("simEnergy signal / Cl. energy")
plt.colorbar(label="Cl. energy [GeV]")
plt.ylim(-0.05, 1.3)
plt.xlim(-0.05, 1.3)
hep.cms.text("Preliminary ",loc=0)
```
