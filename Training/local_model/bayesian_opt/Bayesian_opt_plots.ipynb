```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
import json
from collections import defaultdict

mpl.rcParams['figure.figsize'] = (5,4)
mpl.rcParams['figure.dpi'] = 100
mpl.rcParams["image.origin"] = 'lower'
```

```python
data = defaultdict(list)
for i, l in enumerate(open("logs.json-old")):
    datal = json.loads(l)
    #print(datal)
    data["target"].append(datal["target"])
    data["index"].append(i)
    info = ["batch_size", "dropout", "lr", "l2_reg", "num_layers", "num_units", "optimizer"]
    for i in info:
        data[i].append(datal["params"][i])
```

```python
plt.plot(data["target"])
plt.xlabel("epoch")
plt.ylabel("val_loss")
```

## 1D plots

```python
plt.scatter(data["dropout"], data["target"], c=data["index"])
plt.xlabel("dropout")
plt.ylabel("val_loss")
plt.colorbar(label="iteration")
```

```python
plt.scatter(data["l2_reg"], data["target"], c=data["index"])
plt.xlabel("l2_reg")
plt.ylabel("val_loss")
plt.colorbar(label="iteration")
```

```python
plt.scatter(data["num_units"], data["target"], c=data["index"])
plt.xlabel("N. units")
plt.ylabel("val_loss")
plt.colorbar(label="iteration")
```

```python
plt.scatter(data["num_layers"], data["target"], c=data["index"])
plt.xlabel("N. layers")
plt.ylabel("val_loss")
plt.colorbar(label="iteration")
```

```python
plt.scatter(data["lr"], data["target"], c=data["index"])
plt.xlabel("lr")
plt.ylabel("val_loss")
plt.colorbar(label="iteration")
```

```python
plt.scatter(data["batch_size"], data["target"], c=data["index"])
plt.xlabel("batch_size")
plt.ylabel("val_loss")
plt.colorbar(label="iteration")
```

```python
plt.scatter(data["optimizer"], data["target"], c=data["index"])
plt.xlabel("optimizer")
plt.ylabel("val_loss")
plt.colorbar(label="iteration")
```
